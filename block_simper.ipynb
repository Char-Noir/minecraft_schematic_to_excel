```python
import cv2
import numpy as np
from skimage import io
import os
import json
```

```python
colors ={}
```

```python
def dec_to_hex(numb):
    numb = round(numb)
    return hex(numb)[2:].upper()
```

```python
directory = os.fsencode('block')
    
for file in os.listdir(directory):
     filename = os.fsdecode(file)
     print(filename)
     if filename.endswith(".png"): 
        path = os.path.join('block', filename)
        img = io.imread(path)[:, :, :-1]
        average = img.mean(axis=0).mean(axis=0)
        aver = '#000000'
        if len(average)==3:
         aver = '#' + dec_to_hex(average[0]) + dec_to_hex(average[1]) + dec_to_hex(average[2])
        elif len(average)==2:
         aver = '#' + dec_to_hex(average[0]) + dec_to_hex(average[1]) + '00'
        colors[filename[:-4]] = aver
     else:
         print("not png")
         continue
```

```python
colors
```

```python
new_colors = json.dumps(colors, indent=4)
```

```python
fil = open('colors.json','w')
fil.write(new_colors)
```
